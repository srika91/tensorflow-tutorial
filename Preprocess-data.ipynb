```python
import numpy as np
import scipy
from sklearn.preprocessing import MinMaxScaler,Normalizer,StandardScaler
```

```python
obj = open('heart.txt','r')
data= obj.readlines()
data = data[1:] 
```

```python
content = []
string_feature = []
output_class = []
for i in range(len(data)):
    temp_1 = data[i].split('\t')
    temp = []
    for index in range(len(temp_1)):
        if index== 4 or index == 9:
            temp.append(temp_1[index])
        else:
            temp.append(float(temp_1[index]))
    output_class.append(temp[9])
    string_feature.append(temp[4])
    content.append(temp[:4] + temp[5:9])
print len(content),len(string_feature),len(output_class)
```

```python
content_np = np.array(content)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(content_np)
scaler = StandardScaler().fit(rescaledX)
rescaledX = scaler.transform(rescaledX)
scaler = Normalizer().fit(rescaledX)
normalizedX = scaler.transform(rescaledX)
print normalizedX[:5]
```

```python
final_values = []
myfile = open("heart_preprocessed.csv", 'wb')
for i in range(len(normalizedX)):
    #temp = normalizedX[i,:3].tolist() #+ [string_feature[i]] + normalizedX[i,4:].tolist()+ [output_class[i]]
    temp1 = [str(j) for j in normalizedX[i,:4].tolist()]
    temp2 = [str(j) for j in normalizedX[i,4:].tolist()]
    print string_feature[i]#,output_class[i]
    temp = temp1 + [string_feature[i].replace('"','')] +  temp2 #+ [output_class[i]]
    print temp#,output_class[i]
    for j in temp:
        myfile.write(j+',')
    myfile.write(output_class[i])
```
